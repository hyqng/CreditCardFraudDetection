```python
# Import Libraries
import pandas as pd
from pathlib import Path
```

```python
# Path to raw data folder relative to notebook folder
dataFolder = Path("../data/raw")

# List of CSV files
files = [
    "creditcard_part1.csv",
    "creditcard_part2.csv",
    "creditcard_part3.csv",
    "creditcard_part4.csv"
]

# Read and concatenate all CSVs
dfList = [pd.read_csv(dataFolder / f) for f in files]
allCardsDf = pd.concat(dfList, ignore_index=True)

print(allCardsDf.shape)
print(allCardsDf.head())
```

```python
# Drop rows with any missing values
cleanDf = allCardsDf.dropna()

# Reset the index after dropping rows
cleanDf.reset_index(drop=True, inplace=True)

# Check the result
print(cleanDf.shape)
print(cleanDf.isnull().sum())  # Should be all zeros
```

```python
# Data Integrity and Validation

cleanDf['Time'] = cleanDf['Time'].astype(float)
cleanDf['Amount'] = cleanDf['Amount'].astype(float)

cleanDf['Class'] = cleanDf['Class'].astype(int)

cleanDf.head()
```

```python
# Determine split indices
quarter = len(cleanDf) // 4

# Split into 4 parts
df1 = cleanDf.iloc[:quarter]
df2 = cleanDf.iloc[quarter:2*quarter]
df3 = cleanDf.iloc[2*quarter:3*quarter]
df4 = cleanDf.iloc[3*quarter:]

# Save the four smaller CSVs
df1.to_csv('../data/validated/creditcard_part1.csv', index=False)
df2.to_csv('../data/validated/creditcard_part2.csv', index=False)
df3.to_csv('../data/validated/creditcard_part3.csv', index=False)
df4.to_csv('../data/validated/creditcard_part4.csv', index=False)
```
